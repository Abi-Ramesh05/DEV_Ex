```python
import pandas as pd
import numpy as np
import folium
from IPython.display import display, IFrame
from google.colab import files

# Generate random data
np.random.seed(42)
num_points = 100
latitude = np.random.uniform(37.5, 38.5, num_points)
longitude = np.random.uniform(-123, -121, num_points)
value = np.random.randint(1, 100, num_points)

# Create DataFrame
df = pd.DataFrame({'Latitude': latitude, 'Longitude': longitude, 'Value': value})

# Basic stats
print("ðŸ“Š Basic Statistics:")
print(df.describe())
print("\nMean:", df['Value'].mean())
print("Median:", df['Value'].median())
print("Mode:", df['Value'].mode().values)

# Create the map
base_map = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=10)

for index, row in df.iterrows():
    popup_text = f"Value: {row['Value']}"
    folium.Marker([row['Latitude'], row['Longitude']], popup=popup_text).add_to(base_map)

# Save map to HTML
map_filename = 'interactive_map.html'
base_map.save(map_filename)

print(f"\n Map saved to '{map_filename}'")

# Display map inline
display(base_map)

# Provide download link
files.download(map_filename)
```
